# Matplotlib - Practice Tasks
This notebook contains practice examples for creating and working with matplotlib line chart.

## Line chart
       -A line chart (or line graph) is a type of chart used to visualize data points connected by straight lines.
       -It is commonly used to show trends over time or relationships between variables

**Dataset**:Temperature Changes Over a Week

### Different methods of datatypes
        1.List
        2.Series
        3.Dataframe
        4.Numpy
        5.Function
        6.Dictionary

### 1.Line chart using list

```python
import matplotlib.pyplot as plt
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperature = [26, 33, 30, 29, 31, 26, 27]
plt.plot(days,temperature, color="yellow")
plt.xlabel("days")
plt.ylabel("Temperature")
plt.title("Temperature Change Over a Week")
plt.grid(True)
plt.show()
```

### 2.Line chart using Series

```python
import pandas as pd
```

```python
days = pd.Series(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
temperature =pd.Series ([26, 33, 30, 29, 31, 26, 27])
plt.plot(days,temperature)
plt.xlabel("days")
plt.ylabel("Temperature")
plt.title("Temperature Change Over a Week")
plt.yticks(temperature)
plt.show()
```

### 3.Line chart using Dataframe

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperature = [26, 33, 30, 29, 31, 26, 27]
df= pd.DataFrame({"Daysofweek":days,"Temperatureperday":temperature})
plt.plot(df.Daysofweek,df.Temperatureperday,marker='*',linestyle="dashdot",color="red")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Temperature Change Over a Week")
plt.show()
```

### 4.Line chart using Numpy

```python
import numpy as np
```

```python
temp = np.array([26, 33, 30, 29, 31])
plt.plot(temp,"y")
plt.plot(temp+1,"m")
plt.plot(temp+2,"g")
plt.plot(temp+3,"c")
plt.plot(temp+4,"r")
plt.xlabel("Day Index")
plt.ylabel("Temperature")
plt.title("Temperature Variations with Adjustments")
plt.grid(True)
plt.show()
```

### Different line Designs

```python
temp = np.arange(1,3)
plt.plot(temp,'--',temp+1,'*',temp+2,'--')
plt.xlabel("Day Index")
plt.ylabel("Temperature")
plt.title("Temperature Variations with Adjustments")
plt.show()
```

### 5.Line chart using Function

```python
def fplot(days,temperature):
    plt.plot(days,temperature,color="green")
    plt.xlabel("Days")
    plt.ylabel("Temperature")
    plt.title("Temperature Change Over a Week")
    plt.show()

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperature = [26, 33, 30, 29, 31, 26, 27]
fplot(days,temperature)
```

### 6.Line chart using Dictionary

```python
temp_data = {"Mon": 26, "Tue": 33,"Wed": 30,"Thu": 29,"Fri": 31}

days = list(temp_data.keys())
temperatures = list(temp_data.values())

plt.plot(days, temperatures, marker='o', color='orange', linestyle='-')
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Temperature Change Over a Week")
plt.grid(True)
plt.show()
```

## Multiple Lines in One Chart

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
City1_Temp = [20, 25, 30, 28]
City2_Temp= [15, 18, 22, 26]

plt.plot(months,City1_Temp, label="City1", marker='o')
plt.plot(months,City2_Temp, label="City2", marker='s')

plt.xlabel("months")
plt.ylabel("Temperature")
plt.title("Temperature Change Over a Week")
plt.legend()
plt.show()
```

### Styling and Customization

```python
plt.plot(x,y) 
     #Draws a line chart using x and y data.
plt.xlabel() 
     #Adds a label to the X-axis
plt.ylabel()
     # Adds a label to the Y-axis
plt.xticks()
     # Sets custom tick values or labels on the X-axis
plt.yticks()
     # Sets custom tick values on the Y-axis
plt.legend()
     # Displays the legend for labeled lines
plt.grid(True)
     # Adds grid lines to the chart background
plt.title()
     # Sets the title at the top of the chart
plt.show()
     # Displays the final chart on screen
plt.savefig("linechart.png")
     # Saves the chart as an image file
```

```python
color="red"
      # Sets the line color to red
linestyle="--"
      # Makes the line dashed
marker="o" or "s"
      # Adds circular (`o`) or square (`s`) markers at data points.
markerfacecolor='blue'
      # Fills the marker with blue color
markersize=5
      # Sets the size of each marker to 5
linewidth=3
      # Sets the thickness of the line to 3
```

## Full line Chart with All Styling Options Applied

```python
temp_data={'Mon':26, 'Tue':33, 'Wed':30, 'Thu':29, 'Fri':31, 'Sat':26, 'Sun':27}

days =list(temp_data.keys())
temperature = list(temp_data.values())
plt.plot(days,temperature,label="Temperature changes",marker='o',linestyle="--",color="Orange",markerfacecolor='blue',markersize=5,linewidth=3)
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Temperature Change Over a Week")
plt.xticks(days)
plt.yticks(temperature)
plt.legend()
plt.grid(True)
plt.savefig("linechart.png")
plt.show()
```
